# importing necessary libraries

```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import matplotlib.pyplot as plt
%matplotlib inline
```

# Loading the csv file

```python
data = pd.read_csv('companydata\company_reviews.csv')
```

```python
data.head()
```

# Removing stopwords from the column called Description

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
",".join(stopwords.words('english'))
stop_words=set(stopwords.words('english'))
```

```python
def remove_stop(x):
    return",".join([word for word in str(x).split()if word not in stop_words])
data["description"].head().apply(lambda x : remove_stop(x))
```

# Convert text to lowercase in the description column
we are going to convert the text in Description column into lower case

```python
data["description"]=data["description"].str.lower()
data["description"].head()
```

# Removing special characters from the Locations column

```python
data['locations'] = data['locations'].str.replace('\W', '', regex=True)
data['locations'].head()
```

# Remove URLs from the column "websites"

```python
data['website'] = data['website'].replace(to_replace=r'^https?:\/\/.*[\r\n]*',value='',regex=True)

data['website'].head()
```

# Tokenizing data in the column "description" 

```python
nltk.word_tokenize(data["description"][0])
```
